## ✅ Question 1: Line Charts — Alphabet Inc. Financial Data (Oct 3–7, 2016)

## ✅ Question 3: Line Chart — Year-Wise Sales (monthsales.csv)

```python
# Create monthsales.csv
monthsales_csv = """Month,2018,2019,2020,2021
1,2500,2600,2450,3250
2,2200,1800,2000,3200
3,3000,2500,3000,3800
4,2800,3000,3100,2800
5,2750,3200,1950,2900
6,3500,2000,3550,2300
"""

with open('monthsales.csv', 'w') as f:
    f.write(monthsales_csv)

df_sales = pd.read_csv('monthsales.csv')
print(df_sales)
```

## ✅ Question 6: Scatter Plot + Bar Chart — Population vs Literacy Rate by Region

```python
# Sample Indian state population and literacy data (representative values)
state_data = {
    'State': [
        'West Bengal', 'Odisha', 'Bihar', 'Jharkhand',
        'Rajasthan', 'Gujarat', 'Maharashtra', 'Goa',
        'Uttar Pradesh', 'Punjab', 'Haryana', 'Himachal Pradesh',
        'Tamil Nadu', 'Kerala', 'Karnataka', 'Andhra Pradesh'
    ],
    'Population': [
        91276115, 41974218, 104099452, 32988134,
        68548437, 60439692, 112374333, 1458545,
        199812341, 27743338, 25351462, 6864602,
        72147030, 33406061, 61095297, 49386799
    ],
    'Literacy_Rate': [
        76.26, 72.87, 61.80, 66.41,
        66.11, 78.03, 82.34, 88.70,
        67.70, 75.84, 75.55, 82.80,
        80.09, 93.91, 75.36, 67.02
    ],
    'Region': [
        'East', 'East', 'East', 'East',
        'West', 'West', 'West', 'West',
        'North', 'North', 'North', 'North',
        'South', 'South', 'South', 'South'
    ]
}

df_state = pd.DataFrame(state_data)
print(df_state)
```

```python
# ── Scatter Plot: Population vs Literacy Rate by Region ──
regions = df_state['Region'].unique()
colors  = {'East': 'blue', 'West': 'green', 'North': 'red', 'South': 'orange'}

plt.figure(figsize=(10, 6))
for region in regions:
    subset = df_state[df_state['Region'] == region]
    sizes  = np.sqrt(subset['Literacy_Rate']) * 20
    plt.scatter(subset['Population'], subset['Literacy_Rate'],
                label=region,
                marker='D',
                s=sizes,
                color=colors[region],
                alpha=0.8,
                edgecolors='black')

plt.title('Population vs Literacy Rate by Region', fontsize=14, fontweight='bold')
plt.xlabel('Population')
plt.ylabel('Literacy Rate (%)')
plt.legend(title='Region')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# ── Bar Chart: Average Literacy Rate per Region ──
avg_literacy = df_state.groupby('Region')['Literacy_Rate'].mean().reset_index()
print(avg_literacy)

plt.figure(figsize=(8, 5))
plt.bar(avg_literacy['Region'], avg_literacy['Literacy_Rate'],
        color=['blue', 'red', 'south' and 'orange', 'green'],
        edgecolor='black')

for i, val in enumerate(avg_literacy['Literacy_Rate']):
    plt.text(i, val + 0.3, f'{val:.2f}%', ha='center', fontsize=10)

plt.title('Average Literacy Rate by Region', fontsize=14, fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Average Literacy Rate (%)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Create Marks.csv (extended sample data)
marks_csv = """Name,English,Maths,Hindi,Science,Social_Studies
Rishika Batra,95,95,90,94,95
Waseem Ali,95,76,79,77,89
Kulpreet Singh,78,81,75,76,88
Annie Mathews,88,63,67,77,80
Ravi Sharma,72,85,80,90,74
Priya Gupta,88,92,85,88,91
Amit Verma,65,70,72,68,75
Sonal Joshi,90,88,82,85,87
Deepak Nair,75,80,78,82,79
Meena Iyer,82,77,88,79,83
"""

with open('Marks.csv', 'w') as f:
    f.write(marks_csv)

df_marks = pd.read_csv('Marks.csv')
print(df_marks)
```

```python
subjects = ['English', 'Maths', 'Hindi', 'Science', 'Social_Studies']
data_for_box = [df_marks[subj].values for subj in subjects]

plt.figure(figsize=(10, 6))
bp = plt.boxplot(data_for_box,
                 labels=subjects,
                 patch_artist=True,
                 notch=False)

colors_box = ['#AED6F1', '#A9DFBF', '#F9E79F', '#F1948A', '#D7BDE2']
for patch, color in zip(bp['boxes'], colors_box):
    patch.set_facecolor(color)

plt.title('Subject-wise Performance Analysis (Box Plot)', fontsize=14, fontweight='bold')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Comparative analysis
print('\n📊 Comparative Analysis:')
print(df_marks[subjects].describe().round(2))
```

## ✅ Question 8: Grouped Bar Chart — Scores by Group and Gender

```python
means_men   = (22, 30, 35, 35, 26)
means_women = (25, 32, 30, 35, 29)

n_groups = len(means_men)
index    = np.arange(n_groups)
bar_width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

bars_men   = ax.bar(index,              means_men,   bar_width,
                    label='Men',   color='green', edgecolor='black')
bars_women = ax.bar(index + bar_width, means_women, bar_width,
                    label='Women', color='red',   edgecolor='black')

ax.set_title('Scores by person', fontsize=14, fontweight='bold')
ax.set_xlabel('Person')
ax.set_ylabel('Scores')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels([f'G{i+1}' for i in range(n_groups)])
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

## ✅ Question 9a: Pie Chart — Total Sale Data (Last Year) per Product

```python
# Representative sales data (matching structure of the assignment CSV)
sales_data = {
    'Month': list(range(1, 13)),
    'Bathing_Soap':  [9000, 6700, 9800, 8800, 8000, 7500, 9200, 7800, 8500, 10500, 11500, 13500],
    'Facewash':      [1500, 1200, 1300, 1250, 1750, 1550, 1300, 1450, 1600, 1800,  2100,  1750],
    'Toothpaste':    [5500, 4800, 5200, 5100, 5600, 5400, 5300, 5800, 6000, 6200,  6500,  6800],
    'Shampoo':       [3200, 2900, 3100, 3300, 3500, 3400, 3600, 3700, 3900, 4100,  4200,  4400],
    'Moisturizer':   [2100, 1900, 2000, 2200, 2300, 2150, 2250, 2400, 2500, 2600,  2700,  2800]
}

df_prod = pd.DataFrame(sales_data)

# Save as CSV
df_prod.to_csv('superstore_sales.csv', index=False)
print(df_prod)
```

```python
products   = ['Bathing_Soap', 'Facewash', 'Toothpaste', 'Shampoo', 'Moisturizer']
total_sales = df_prod[products].sum()

explode = [0.05] * len(products)

plt.figure(figsize=(9, 7))
plt.pie(total_sales,
        labels=products,
        autopct='%1.1f%%',
        explode=explode,
        startangle=140,
        shadow=True)

plt.title('Total Units Sold per Product (Annual)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nTotal units sold per product:')
for p, v in zip(products, total_sales):
    print(f'  {p}: {v}')
```

## ✅ Question 10: Histogram — Total Monthly Profit

```python
# Total monthly profit (computed from sales data as a demonstration)
profit_per_unit = {
    'Bathing_Soap': 12, 'Facewash': 45,
    'Toothpaste': 18,   'Shampoo': 22,
    'Moisturizer': 35
}

df_prod['Total_Profit'] = sum(
    df_prod[p] * profit_per_unit[p] for p in products
)

print('Monthly Total Profit:')
print(df_prod[['Month', 'Total_Profit']])
```

```python
profit_values = df_prod['Total_Profit'].values

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(profit_values,
                             bins=6,
                             color='steelblue',
                             edgecolor='black',
                             label='Profit data')

plt.title('Profit data', fontsize=14, fontweight='bold')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print(f'\nProfit range: {profit_values.min():,} — {profit_values.max():,}')
print(f'Most common profit range: {bins[np.argmax(n)]:,.0f} — {bins[np.argmax(n)+1]:,.0f}')
```
